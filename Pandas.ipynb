QUESTION 1) Update items of list “a” into list “b”. Items in list b should not be duplicated.

a = [ {'name': 'a1' }, {'name': 'a2' }, {'name': 'a3' }, {'name': 'a4' }, {'name': 'a5' } ]

b = [ { 'name': 'a3' }, {'name': 'a5' } ]

Output ⇒ b = [ { 'name': 'a3' }, {'name': 'a5' }, {'name': 'a1' }, {'name': 'a2' }, {'name': 'a4' } ]

```python
a = [ {'name': 'a1' }, {'name': 'a2' }, {'name': 'a3' }, {'name': 'a4' }, {'name': 'a5' } ]
b = [ { 'name': 'a3' }, {'name': 'a5' } ]
result=[]
for item_b in b:
  if item_b not in result:
    result.append(item_b)

for item_a in a:
  if item_a not in result:
    result.append(item_a)

result.sort(key=lambda x: x['name'])
print(result)
```
